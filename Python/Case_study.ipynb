```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Next the x, y and z-axis values are calculated using the numpy Library. 

```python
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
```

Finally the Matplotlib library is used to plot a graph.

```python
plt.plot(x,y,x,z)
plt.xlabel('x in radians from 0 to 4$\pi$')  # string must be enclosed with quotes '  '
plt.ylabel('sin(x) and cos(x)')
plt.title(title)
plt.legend(['sin(x)', 'cos(x)'])      # legend entries as seperate strings in a list
plt.grid(axis = 'y')
plt.show()
```
